Na aula nós realizamos a normalização em um arquivo com o formato JSON. No entanto, a função json_normalize() aceita apenas um dicionário ou uma lista de dicionários. Por isso, no vídeo foi usada a estratégia de usar o código: pd.json_normalize(dados_pacientes_2['Pacientes']). Porém, se tentarmos usar parâmetros da função json_normalize em um arquivo JSON podem surgir erros.
```python
#Importando a biblioteca Pandas
import pandas as pd

#Importando o módulo JSON
import json

#Lendo o arquivo json usando o módulo Python JSON
with open('pacientes_2.json','r') as f:
    dados = json.loads(f.read())

#Normalizando os dados com os parâmetros record_path e meta
pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
```

```python
import sqlalchemy
```

Também importamos a classe Table que representa uma tabela em um banco de dados e é usada para manipular dados em uma tabela específica.

Por fim, importamos a função inspect, que permite inspecionar a estrutura do banco de dados.

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

Criamos o banco de dados. Para isso, criamos uma variável chamada engine. Dela, chamamos o create_engine(). Dentro das chaves abrimos aspas simples e escrevemos o banco de dados que será utilizado, 'sqlite:///:memory:'.

Nesse caso, usaremos o SQLite, um banco de dados que vem naturalmente instalado no Google Colab. Nesse trecho de código o memory indica que estamos usando um banco de dados local.

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

Quando trabalhamos com SQL utilizamos cláusulas, ou seja, palavras que possuem funções específicas para a consulta.

Lembrando que em SQL as cláusulas são escritas em caixa alta.

Sendo assim, após as aspas, escrevemos SELECT seguido de *. Esse asterisco indica que queremos selecionar todas as colunas da tabela.

Agora, precisamos especificar qual tabela é essa. Então, após * inserimos a cláusula FROM clientes.

Na mesma linha, escrevemos a cláusula WHERE, que filtra as informações, seguido do nome da coluna que queremos consultar Categoria_de_renda. Logo após escrevemos ="Empregado" e executamos.

**Estamos pedindo para selecionar todas as colunas da tabela clientes na qual a categoria de renda seja igual a "Empregado".**

Para ler uma consulta em SQL criaremos a variável empregados = pd.read_sql(). Dentro do parêntese, passamos a consulta e o motor do bando de dados, query, engine. Em seguida, a executamos.

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

Para salvá-la como uma nova tabela escrevemos empregados.to_sql(). Dentro do parêntese inserimos aspas simples e dentro dela escrevemos empregados. Adicionamos uma vírgula e passamos o parâmetro con=engine seguido de index=False e executamos.

```python
empregados.to_sql('empregados',con=engine, index=False)
```

Utilizamos o read_sql para lermos uma consulta, mas e se quisermos ler uma tabela inteira como a que acabamos de salvar?

Para isso, utilizamos a função pd.read_sql_table(). Dentro do parêntese inserimos aspas simples, seguido do nome da tabela que queremos executar 'empregados' e engine. Executamos em sequência.

**Lembre-se sempre dessa diferença, o read_sql permite ler uma consulta e o read_sql_table uma tabela completa**

```python
pd.read_sql_table('empregados', engine)
```

**Deletando item do banco de dados**

O cliente 5008808 está com o grau de escolaridade como Ensino médio. Porém, ele fez faculdade e deseja mudar para Ensino superior.

Para isso, em uma nova linha escrevemos a consulta query =''. Dentro das aspas usamos a clausula UPDATE seguido do nome da tabela clientes e a clausula SET.

Após, definimos qual coluna será alterada, então Grau_escolaridade. Inserimos o símbolo de igual, seguido da informação que queremos atualizar entre aspas duplas, nesse caso "Ensino superior".

Definimos onde isso será alterado WHERE ID_Cliente. Inserimos o símbolo de igual e o ID do cliente 5008808.

Agora, copiamos o mesmo código do contexto with que usamos anteriormente e colamos abaixo dessa nova consulta. Em seguida, executamos a célula. Dessa forma:

```python
# Deletando itens do banco de dados
query = 'DELETE FROM clientes WHERE ID_Cliente="5008804"'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```

**Atualizando um banco de dados**

Suponhamos que a pessoa de ID 5008804 cancelou sua conta na instituição financeira. Isso significa que ela não faz mais parte do quadro de clientes.

Para tirá-la da tabela precisamos realizar uma consulta. Então, escrevemos query = ''. Dentro das aspas, inserimos a instrução DELETEseguido da cláusula FROM clientes. Agora, especificamos qual cliente será deletado, então, escrevemos WHERE seguido do nome da coluna ID_Cliente e o número do ID 5008804.

Apertamos "Enter" para a linha de baixo. Agora, fazemos um contexto with. Isso garante que a conexão com o banco de dados seja fechada corretamente após a execução da consulta. Depois, usamos o método engine.connect(), seguido de as conn.

Na linha de baixo, chamamos o conn.execute(query). Em seguida, executamos.

Com isso estamos usando um método chamado connect para estabelecer uma conexão com o banco de dados. Estamos pedindo para executar a query com o método execute que quando chamado, executa a instrução SQL diretamente no banco de dados, usando a conexão estabelecida.

```python
# Atualizando itens do banco de dados
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente="5008808"'
with engine.connect() as conn:
  conn.execute(query)
```
